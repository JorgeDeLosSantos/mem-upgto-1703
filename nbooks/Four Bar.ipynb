```python
from sympy import init_printing, solve, symbols, Eq
init_printing(use_latex=True)
A,B,C,D,E,F,x,y = symbols('A,B,C,D,E,F,x,y')

eq1 = Eq(A*x + B*y, C)
eq2 = Eq(D*x + E*y, F)
solve([eq1,eq2],[x,y])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi,cos,sin,tan,sqrt,arcsin,arccos,arctan2
%matplotlib inline

def plotvector(p0,u,*args,**kwargs):
    """
    Gráfica un vector, dados su punto inicial 
    y sus componentes rectangulares
    """
    plt.plot([p0[0],p0[0]+u[0]],[p0[1],p0[1]+u[1]],*args,**kwargs)


for t2 in np.linspace(0,pi,5):
    r1,r2,r3,r4 = 7,3,8,6
    t1 = 0
    z = sqrt(r1**2 + r2**2 - 2*r1*r2*cos(t2))
    alpha = arcsin((r2/z)*sin(t2))
    beta = arccos(-(r3**2 - r4**2 - z**2)/(2*r4*z))
    gamma = arcsin((z/r3)*sin(beta))
    t3 = pi - (alpha+beta+gamma)
    t4 = pi - (alpha+beta)

    R1 = np.array([r1*cos(t1),r1*sin(t1)])
    R2 = np.array([r2*cos(t2),r2*sin(t2)])
    R3 = np.array([r3*cos(t3),r3*sin(t3)])
    R4 = np.array([r4*cos(t4),r4*sin(t4)])
    
    clr = np.array([t2,t2,t2])/pi
    plotvector([0,0], R2, color=clr)
    plotvector(R2, R3, color=clr)
    plotvector(R1, R4, color=clr)
    plt.axis('equal')
    plt.grid(ls="--")
```
