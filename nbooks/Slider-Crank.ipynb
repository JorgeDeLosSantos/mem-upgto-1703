## Mecanismo de manivela-biela corredera

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi,cos,sin,tan
%matplotlib inline

def plotvector(p0,u,color="r"):
    """
    Gr√°fica un vector, dados su punto inicial 
    y sus componentes rectangulares
    """
    plt.plot([p0[0],p0[0]+u[0]],[p0[1],p0[1]+u[1]], color=color, ls="-", marker='o')

r2, r3 = 3, 8
for t2 in np.linspace(0,2*pi,10):
    color = np.array([t2,0,t2])/(2*pi)
    t3 = np.arcsin(-r2*sin(t2)/r3)
    r1 = r2*cos(t2) + r3*cos(t3)
    R2 = np.array([r2*cos(t2),r2*sin(t2)])
    R3 = np.array([r3*cos(t3),r3*sin(t3)])
    plotvector([0,0], R2, color)
    plotvector(R2, R3, color)
    plt.grid('on')
    plt.axis('equal');
```
